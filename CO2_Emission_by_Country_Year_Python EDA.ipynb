```python
def filter_country(search_term):
    '''
    Filters the medals_df DataFrame to only contain rows whose Sport or Discipline
    columns contain the given search_term.
    '''
    return mydf[mydf["Country Name"].str.contains(search_term, case=False) \
                     | mydf["Country Code"].str.contains(search_term, case=False)\
                     | mydf["Region"].str.contains(search_term, case=False)]
```

```python
d1 = filter_country("Japan")
d2 = filter_country("Australia")
d3 = filter_country("United States")
d4 = filter_country("China")
d5 = filter_country("United Kingdom")
d6 = filter_country("Saudi Arabia")
df_set = pd.concat([d1, d2, d3, d4, d5, d6])
df_set.reset_index(drop=True)
```

```python
dfnew = df_set.drop(["Country Name","TableName","Country Code","SpecialNotes","Region"],axis=1)
dfnew.drop(dfnew.columns[dfnew.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
dfnew
```

```python
import seaborn as sns
dfnew = dfnew.set_index('IncomeGroup')
dfnew = dfnew.stack()
dfnew.index = dfnew.index.rename('year', level=1)
dfnew.name = 'CO2'
dfnew = dfnew.reset_index()
dfnew
```

```python
ax = sns.boxplot(x="IncomeGroup", y="CO2", data=dfnew)
vals = dfnew
plt.xticks(range(len(range(2,4))),dfnew["IncomeGroup"].unique())
plt.title('CO2 Emission by Incme Group', fontsize=24)
plt.savefig('Boxplot Grapgh.png')
```
